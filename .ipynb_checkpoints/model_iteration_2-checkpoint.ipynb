#### Fare

The test data lacks a single Fare variable. To compensate for this, I will find the mean for the test fair and replace the missing value with the mean. Also, I will alter the Fare values in the dataset to integers to make processing easier.

As seen in the plot of Fare against Survival below, many people who paid higher fares seemed to have survived the Titanic more. The variable seems significant and will remain for the machine learning algorithm.

```python
# only for test_df, since there is a missing "Fare" values
test["Fare"].fillna(test["Fare"].mean(), inplace=True)

# convert from float to int
titanic['Fare'] = titanic['Fare'].astype(int)
test['Fare']    = test['Fare'].astype(int)

# Plotting Fare against survival rate to see if it had an impact
sns.lmplot("Fare", "Survived", titanic, x_bins=4, logistic=True, truncate=True);
```

#### Age

The next variable that I will deal with is Age. Age has a lot of missing values in both datasets, and it is a seemingly dire factor in survival, so filling out the values hold great importance. I used two different methods to fill the empty age values. One method is from Source 1, and the other is from Source 4. 

Source 1 just replaces the missing values with the average, while Source 4 replaces the missing values with a random combination based on mean and standard deviation of the age values. The method of Source 4 seems a bit superior as it adds variation and may prevent overfitting to the training set.

```python
# Source 1 Method
#titanic["Age"] = titanic["Age"].fillna(titanic["Age"].median())
#test["Age]=test["Age"].fillna(test["Age"].median())

# Source 4 Method
# get average, std, and number of NaN values in titanic_df
average_age_titanic   = titanic["Age"].mean()
std_age_titanic       = titanic["Age"].std()
count_nan_age_titanic = titanic["Age"].isnull().sum()

# get average, std, and number of NaN values in test_df
average_age_test   = test["Age"].mean()
std_age_test       = test["Age"].std()
count_nan_age_test = test["Age"].isnull().sum()

# generate random numbers between (mean - std) & (mean + std)
rand_1 = np.random.randint(average_age_titanic - std_age_titanic, average_age_titanic + std_age_titanic, size = count_nan_age_titanic)
rand_2 = np.random.randint(average_age_test - std_age_test, average_age_test + std_age_test, size = count_nan_age_test)

# fill NaN values in Age column with random values generated
titanic["Age"][np.isnan(titanic["Age"])] = rand_1
test["Age"][np.isnan(test["Age"])] = rand_2

# convert from float to int
titanic['Age'] = titanic['Age'].astype(int)
test['Age']    = test['Age'].astype(int)

# average survived passengers by age
fig, axis1 = plt.subplots(1,1,figsize=(18,4))
average_age = titanic[["Age", "Survived"]].groupby(['Age'],as_index=False).mean()
sns.barplot(x='Age', y='Survived', data=average_age)
```

Looking at the distribution between age and survival, the relationship between age and survival seems parabolic rather than linear. Also considering the drastic changes in survival rates between different genders in the same age group, it may be wise to create new categories that tie age with gender rather than run those two categories as separate variables.

#### Family Members

Next, we look at the companionship of each member on the Titanic. Survival rates logically seem to change as an individuals had family members on a ship as there were people looking after each other. Source 2 and Source 4 implement this consideration, although Source 2 does it in an analog method where it counts numbers of family members, while Source 4 discretely mentions if a person had family on the ship or not. Source 2 also implements a family id feature that groups people together who are in the same family.

Both methods implement the categories of Sibsp and Parch. However, I am leaning towards the method of Source 4, as having more family members increase the odds of surviving overall, women with greater members tend to be left behind, as shown in the graphs below.

```python
# Creating New category Alone for creating graph. 
#This will be deleted later as it wont help with regression
titanic["alone"] = ~(titanic.Parch + titanic.SibSp).astype(bool)
titanic["Fam_tot"] = titanic.Parch+ titanic.SibSp
# Graphing two graphs for survival rates based on family, one for general, one by gender
sns.factorplot("alone", "Survived", data=titanic).set(ylim=(0, 1));
sns.lmplot("Fam_tot", "Survived", titanic, hue="Sex", x_estimator=np.mean, logistic=True, palette=pal);

# Dropping Variables after graphing process
titanic = titanic.drop(['alone','Fam_tot'], axis=1)

```

```python
# Source 4 Method, Discrete Family member count

# We create a new variable family that adds two values of parch and sibsp together
titanic['Family'] =  titanic["Parch"] + titanic["SibSp"]

# We make a boolean(ish) statement to simplify the analog variable to discrete
titanic['Family'].loc[titanic['Family'] > 0] = 1
titanic['Family'].loc[titanic['Family'] == 0] = 0

#We do the same for the test data set
test['Family'] =  test["Parch"] + test["SibSp"]
test['Family'].loc[test['Family'] > 0] = 1
test['Family'].loc[test['Family'] == 0] = 0

# drop Parch & SibSp from the data set
titanic = titanic.drop(['SibSp','Parch'], axis=1)
test    = test.drop(['SibSp','Parch'], axis=1)
```

#### Age and Gender

Age seems to have a non linear, parabolic relationship with survival. Combine this with the variation in survival rates in the genders, and the significance of age plummets significantly. In source 2 (dataquest: improving the model), Age has a very low score as a predictor due to the combination with gender. Therefore, based upon Source 4's method, I created a new category called "Person" that is defined by age and sex. The three categories are Men, Women, and Children. Although I wanted to create 4 groups (including gender divisions within the children group), I did not know how to create a dummy variable with 4 options.

I considered dropping Age and ran the tests with and without the age value. Although the coefficient of age is very low in Source 2 (dataquest: improvement), it seems to have a significant benefit to the random forest regression. I believe that is because of the parabolic trend of how younger and older people still seem to survive more, despite the chaos caused by different genders in the same age group.

```python
# Function that takes in Age and Sex and defines group as Male, Female, or Child
def get_person(passenger):
    age,sex = passenger
    return 'child' if age < 16 else sex

# Adding new category "Person" to Data Set, using get_person function
titanic['Person'] = titanic[['Age','Sex']].apply(get_person,axis=1)
test['Person']    = test[['Age','Sex']].apply(get_person,axis=1)

# Drop Sex and Age column as we defined it with "Person" Column
titanic.drop(['Sex'],axis=1,inplace=True)
test.drop(['Sex'],axis=1,inplace=True)

# create dummy variables for Person column, & drop Male as it has the lowest average of survived passengers
person_dummies_titanic  = pd.get_dummies(titanic['Person'])
person_dummies_titanic.columns = ['Male','Female','Child']
person_dummies_titanic.drop(['Male'], axis=1, inplace=True)

person_dummies_test  = pd.get_dummies(test['Person'])
person_dummies_test.columns = ['Male','Female','Child']
person_dummies_test.drop(['Male'], axis=1, inplace=True)

titanic = titanic.join(person_dummies_titanic)
test    = test.join(person_dummies_test)
```

#### Title

As seen in Source 2, the title of a passenger plays a crucial role in survival. Richer people may have certain titles more often than not. Based on the code in Source 2, I filter through the names of the passenger and find their title and create a new "Title" category. We know that the title is very important as seen in the graph in Source 2 that the accuracy score of the Title category was very high.

Also in the plot below, each individual category (or different title group) has very distinct survival rates from one another. The Key is the print of the dictionary where each title type correlates with the classification number. 


I am also getting rid of the Titles category because although initially I really wanted to keep it in my regression, the Title is in the form of an object, and cannot be processed by the regression method used below. Therefore, I will have to drop the title. On the bright side, Fare seemed to have high correlation with title, and not much will be lost.

```python
# A function to get the title from a name. From Source 2
def get_title(name):
    # Use a regular expression to search for a title.  Titles always consist of capital and lowercase letters, and end with a period.
    title_search = re.search(' ([A-Za-z]+)\.', name)
    # If the title exists, extract and return it.
    if title_search:
        return title_search.group(1)
    return ""

# Get all the titles and print how often each one occurs.
titles = titanic["Name"].apply(get_title)
titles_test = test["Name"].apply(get_title)

# Map each title to an integer.  Some titles are very rare, and are compressed into the same codes as other titles.
title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Dr": 5, "Rev": 6, "Major": 7, "Col": 7, "Mlle": 8, "Mme": 8, "Don": 9, "Lady": 10, "Countess": 10, "Jonkheer": 10, "Sir": 9, "Capt": 7, "Ms": 2}
for k,v in title_mapping.items():
    titles[titles == k] = v
for k,v in title_mapping.items():
    titles_test[titles_test == k] = v

# Add in the title column.
titanic["Title"] = titles
test["Title"] = titles_test
```

```python
# Graphing different Titles against Survival Rate
sns.factorplot("Title", "Survived", data=titanic, palette="PuBu_d", join=False);
titanic.drop("Title",axis=1,inplace=True)
test.drop("Title",axis=1,inplace=True)
```

#### Cabin Class

The cabin class relates directly to survival, as it is obvious by the graph below. However, rather than using the cabin class category as a analog variable that scales from 1 to 3, using a dummy variable with just cabins 1 and 2 (as cabin 3 has many people die and doesnt aid in survival) may improve the influence of the cabin category in the test.

```python
# sns.factorplot('Pclass',data=titanic_df,kind='count',order=[1,2,3])
sns.factorplot('Pclass','Survived',order=[1,2,3], data=titanic,size=5)

# Dummy variable created for Class in both test and training data.
# Class 3 is omitted as lowest survival rate and acts as control.
pclass_dummies_titanic  = pd.get_dummies(titanic['Pclass'])
pclass_dummies_titanic.columns = ['Class_1','Class_2','Class_3']
pclass_dummies_titanic.drop(['Class_3'], axis=1, inplace=True)

pclass_dummies_test  = pd.get_dummies(test['Pclass'])
pclass_dummies_test.columns = ['Class_1','Class_2','Class_3']
pclass_dummies_test.drop(['Class_3'], axis=1, inplace=True)

# Dropping Pclass category as redundant with soon to be inputted dummie variable
titanic.drop(['Pclass'],axis=1,inplace=True)
test.drop(['Pclass'],axis=1,inplace=True)

# Add dummie variables to categories
titanic = titanic.join(pclass_dummies_titanic)
test    = test.join(pclass_dummies_test)
```

Due to the fact that the testing process cannot take in strings, we will have to cut out some categories to prepare the training and testing dataset. For the training data set, we will eliminate name and person columns, while in the test dataset we will eliminate persons.

### Step 3. Testing

Now, to test the efficiency of the model, I will divide up the training data set into different folds and test the algorithms against one another. 

```python
# dividing up data into different sets

X_train = titanic.drop("Survived",axis=1)
Y_train = titanic["Survived"]
X_test  = test.drop("PassengerId",axis=1).copy()
```

#### Logistic Regression

Now with the divided data sets, I will apply logistic regression that is heavily used in Source 1 and 2. 

```python
# Logistic Regression applied to Data Sets

log_reg = LogisticRegression()
log_reg.fit(X_train, Y_train)
Y_pred = log_reg.predict(X_test)
log_reg.score(X_train, Y_train)
```

#### Random Forest Method

The Logistic regression gives us an accuracy of about 80.6%. We will now attempt the new method introduced in Source 2, and utilized in Source 4, which is the Random Forest Method.

```python
# Random Forests Method
# We set the number of estimators at 100 to give many regressions to work off of
# Important not to make the number of n too big as it may overfit.

random_forest = RandomForestClassifier(n_estimators=50)
random_forest.fit(X_train, Y_train)
Y_pred = random_forest.predict(X_test)
random_forest.score(X_train, Y_train)
```

The Random Forest method with 100 iterations offers a whopping 96% accuracy on the training data. Adding back the age into the categories, which I initially thought was very insignificant due to the lack of score in Source 2's graph, actually increased theh score a lot. This shows how the computation methods of the Logistic Regression and Random Forrest method have somewhat of a difference with each other.
